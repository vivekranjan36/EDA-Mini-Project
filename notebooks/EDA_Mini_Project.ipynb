fifa.csv – data file.

fifa_ variable_information.csv - information on individual variables.

# Dropping Unnecessary Columns

```python
df_fifa=df1.drop(['Photo','Club Logo','Jersey Number','Loaned From','Flag'],axis=1)
df_fifa.head()
```

```python
df_fifa._get_numeric_data().head()
```

# Data Cleaning :

# Value and Wage Columns 

```python
def extract_values(Value):
    i = Value.replace('€','')
    if 'M' in i:
        i = float(i.replace('M', ''))*1000000
    elif 'K' in i:
        i = float(i.replace('K', ''))*1000
    return float(i)

df_fifa['Value1'] = df_fifa['Value'].apply(lambda x: extract_values(x))
df_fifa['Wage1'] = df_fifa['Wage'].apply(lambda x: extract_values(x))

print(df_fifa['Value1'].head())
df_fifa['Value1'].isnull().any()
```

# Joined Column

```python
df_fifa['Join_year'] = df_fifa['Joined'].dropna().map(lambda x: x.split(' ')[2])
df_fifa['Join_year'].head()
```

# Contract Valid Until

```python
df_fifa['Contract Valid Until'] =  pd.to_datetime(df_fifa['Contract Valid Until'])
df_fifa['Contract Valid Until'].head()
```

# Height Column

```python
df_fifa['New Height'] = df_fifa['Height'].dropna().map(lambda x: x.split("'")).str.join(".").astype(float)
df_fifa['New Height'].head()
```

# Weight Column

```python
df_fifa['New Weight']=df_fifa['Weight'].dropna().map(lambda x: x.strip('lbs')).astype(float)
df_fifa['New Weight'].head()
```

# Release Clause Column 

```python
def extract_values(a):
    i = a.replace('€','')
    if 'M' in i:
        i = float(i.replace('M', ''))*1000000
    elif 'K' in i:
        i = float(i.replace('K', ''))*1000
    return float(i)

df_fifa['New_Release Clause'] = df_fifa['Release Clause'].dropna().apply(lambda x: extract_values(x))

df_fifa['New_Release Clause'].head()
```

# Dropping the old columns which have been converted

```python
df_fifa2 = df_fifa.drop(['Value','Wage','Joined','Height','Weight','Release Clause'], axis=1)
df_fifa2.head()
```

# Imputing Categorical values by Mode

```python
df_miss = pd.DataFrame(df_fifa2.isnull().any())
print(list(df_miss[df_miss[0]==True].index))
mode1=df_fifa2['Club'].mode()
mode2=df_fifa2['Preferred Foot'].mode()
mode3=df_fifa2['Work Rate'].mode()
mode4=df_fifa2['Body Type'].mode()
mode5=df_fifa2['Position'].mode()
mode6=df_fifa2['Contract Valid Until'].mode()
mode7=df_fifa2['Join_year'].mode()
```
